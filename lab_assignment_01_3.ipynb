```python
new_lowercase_text = new_sample_text.lower()
print(f"Original text: {new_sample_text}")
print(f"Lowercase text: {new_lowercase_text}")
```

Counting words in the new text using NLTK:

```python
from nltk.tokenize import word_tokenize

new_nltk_words = word_tokenize(new_sample_text)
new_word_count_nltk = len(new_nltk_words)

print(f"Word count (NLTK) for new text: {new_word_count_nltk}")
```

Counting words in the new text using spaCy:

```python
# Process the new text with spaCy
new_doc = nlp(new_sample_text)

# Count tokens that are not punctuation or whitespace
new_spacy_words = [token.text for token in new_doc if not token.is_punct and not token.is_space]
new_word_count_spacy = len(new_spacy_words)

print(f"Word count (spaCy) for new text: {new_word_count_spacy}")
```
